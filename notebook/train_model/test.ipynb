```python
def augment_batch_gemini(text, n=3):
    model_id = "gemini-2.0-flash"
    # One request for multiple results
    prompt = f"Rewrite this Khmer sentence in {n} different ways. Return ONLY a bulleted list: '{text}'"
    
    try:
        response = client.models.generate_content(model=model_id, contents=prompt)
        # Split by newline and clean up bullet points/numbers
        return [line.strip('*- •123456789. ') for line in response.text.strip().split('\n') if line.strip()]
    except Exception as e:
        print(f"Error: {e}")
        return []

# Usage
new_texts = augment_batch_gemini("ខ្ញុំ សុខសប្បាយ", n=3)
```
