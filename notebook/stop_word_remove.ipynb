```python
import pandas as pd

# Load the stopword CSV
stopwords_df = pd.read_csv('../stop_word/khmer stopwords-corpus-385.csv', header = None)  # adjust filename
print(stopwords_df.head())
```

```python
# Remove leading/trailing whitespace, tabs, and carriage returns
stopwords_df[0] = stopwords_df[0].str.strip()

# Check the result
print(stopwords_df.head())
```

```python
stopwords_df
```

```python
stopwords_list = stopwords_df[stopwords_df.columns[0]].tolist()
```

```python
stopwords_list
```

```python
# Remove zero-width spaces and extra whitespace
stopwords_list = [word.replace('\u200b', '').strip() for word in stopwords_list]

# Optional: remove empty strings
stopwords_list = [word for word in stopwords_list if word]
```

```python
stopwords_list
```

```python
# Words you want to remove from stopwords
words_to_remove = ['ទេ', 'គាំទ្រ', 'ច្រើន', 'ច្រើនជាងគេ', 'មិនដែល', 'មិនត្រូវ', 'មិនទាន់', 'មិនទៀង', 'មិនព្រម', 'មិនមែន', 'មិនអាច',]

# Remove them from stopwords_list
stopwords_list = [word for word in stopwords_list if word not in words_to_remove]

# Check the result
print(stopwords_list[:10])  # show first 10 words
```

```python
stopwords_list
```

```python
import pandas as pd

df = pd.DataFrame(stopwords_list, columns=['stopword'])
df.to_csv('../stop_word/khmer_stopwords.csv', index=False, encoding='utf-8-sig')
```
